#Rectified Linear(ReLU) OBJECT


*   X (capital X) - features set in ml
*   y (small y) - Targets or the classification



```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data

nnfs.init()

X = [[1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]]

X, y = spiral_data(100, 3) #States 100 features states of 3 classes

class Layer_Dense:
  def __init__(self, n_inputs, n_neurons):
    self.weights = 0.10 * np.random.randn(n_inputs, n_neurons)  #multiplying by 0.10 to get values betwwen -1 and 1
    self.biases = np.zeros((1, n_neurons))    # 1- is the shape of bias and also no need for Transpose as np.np.random.randn handles that
  def forward(self, inputs):
    self.output = np.dot(inputs, self.weights) + self.biases

#Rectified Linear(ReLU) OBJECT
class Activation_ReLU:
  def forward(self, inputs):
    self.output = np.maximum(0, inputs)

layer1 = Layer_Dense(2 , 5)   #(2- no. of features per input/sample(here spirial dataset has x and y axis), 5- no of neurons we nedd(LITERALLY CAN BE ANY VALUE))
activation1 = Activation_ReLU()

layer1.forward(X)  #calculation done here (input*weights)+bias

#print(layer1.output) #values before activation fn (has many -ve values afetr calculation above)

activation1.forward(layer1.output)  #We are inputing the calulatedd valuess in ReLU to make -ve values to 0 and +ve values remains same
print(activation1.output)
```

```python
import numpy as np

X = [[1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]]

inputs = [0, 2, -1, 3.3, -2.7, 1.1, 2.2, -100]
output = []

# Below is the ReLU activation fn,
# ReLU activation fn = Everything below zero is Clipped to 0 and evrything above 0 remains same
for i in inputs:
    if i > 0:
      output.append(i)
    elif i <= 0:
      output.append(0)

print(output)
```

Why ReLU often beats sigmoid?

ReLU (Rectified Linear Unit) has become the default activation for most hidden layers because it addresses several practical shortcomings of the sigmoid:




*   Avoids vanishing gradients – For positive inputs the derivative is 1, so gradients stay strong during back‑propagation, whereas sigmoid’s derivative shrinks to 0 at both ends, slowing learning in deep nets.
*   Computationally cheap(FAST) – ReLU only requires a max‑operation (ReLU(x)=max⁡(0,x)ReLU(x)=max(0,x)), while sigmoid needs an exponential, making forward and backward passes faster.



*   Promotes sparse activations – Negative inputs output exactly 0, leading to fewer active neurons and a form of built‑in regularization that often improves generalization.
