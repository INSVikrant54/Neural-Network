```python
import numpy as np

X = [[1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]]

class Layer_Dense:
  def __init__(self, n_inputs, n_neurons):
    self.weights = 0.10 * np.random.randn(n_inputs, n_neurons)  #multiplying by 0.10 to get values betwwen -1 and 1
    self.biases = np.zeros((1, n_neurons))    # 1- is the shape of bias and also no need for Transpose as np.np.random.randn handles that
  def forward(self, inputs):
    self.output = np.dot(inputs, self.weights) + self.biases

layer1 = Layer_Dense(4 , 5)   #(4- size of shape provided by input, 5- no of neurons we nedd(LITERALLY CAN BE ANY VALUE))
layer2 = Layer_Dense(5 , 2)  #(5- no. neurons output from previous layer, 2- no of neurons we need can be any)

layer1.forward(X)
#print(layer1.output)
layer2.forward(layer1.output)
print(layer2.output)
```



*   ***def __init__(self)***:: This is a special method called a constructor. It runs automatically when you create a new instance of the class.

*   **self:** This is a required first parameter in class methods. It refers to the specific instance of the object being created, allowing it to keep its own unique data (like its own specific weights and biases).


*   **def forward(self)**:: This is a regular instance method. In deep learning, we name it forward because it represents the "forward pass" where data flows through the layer to produce an output.
*   **pass**: This is a null statement. Python uses it as a placeholder when a statement is syntactically required but you haven't written the logic yet. It literally tells Python to "do nothing."
