```python
import math
softmax_output = [0.7, 0.1, 0.2]

#target_class = 0
target_output = [1, 0, 0]

loss = -(math.log(softmax_output[0])*target_output[0] +
         math.log(softmax_output[1])*target_output[2] +
         math.log(softmax_output[2])*target_output[2] )

print(loss)
loss = -math.log(softmax_output[0])  #obtaining same value as above (just simplified as rest valuess are just 0 and 1)
print(loss)

```
